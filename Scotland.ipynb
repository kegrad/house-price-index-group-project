```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Average-prices-2022-07.csv")
```

```python
df.head()
```

```python
df_Date = df.loc[df.Date == '2022-07-01',:].copy()
df_Date.head()
```

```python
df_scotland = df[df['Area_Code'].str.match('S' )]
df_scotland = df_scotland[df_scotland['Region_Name'] != 'Scotland']
df_scotland
```

```python
plt.figure(figsize=(16, 8), dpi=300)
df_averag_price=df_scotland.pivot(index='Date',columns='Region_Name',values='Average_Price')
df_averag_price[['Highland','Stirling','Orkney Islands']].plot()
# adding title to the plot
plt.title('')
  
# adding labels to the axes, add a legend
plt.xlabel('Date')
plt.ylabel('Average_Price')
plt.legend()
```

```python
plt.figure(figsize=(16, 8), dpi=300)
df_annual_change=df_scotland.pivot(index='Date',columns='Region_Name',values='Annual_Change')
df_annual_change[['Highland','Stirling','Orkney Islands']].plot(label='Annual_Change', kind='bar')
# adding title to the plot
plt.title('')
  
# adding labels to the axes, add a legend
plt.xlabel('Date')
plt.ylabel('Annual_Change')
plt.legend()
```
