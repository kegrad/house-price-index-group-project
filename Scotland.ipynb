```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Average-prices-2022-07.csv")
```

```python
#df.head()
##df_Date = df.loc[df.Date == '2022-07-01',:].copy()
##df_Date.head()
```

```python
df_scotland = df[df['Area_Code'].str.match('S' )]
df_scotland = df_scotland[df_scotland['Region_Name'] != 'Scotland']
#df_scotland.head(50)
```

```python
plt.figure(figsize=(16, 8), dpi=300)
df_average_price=df_scotland.pivot(index='Date',columns='Region_Name',values='Average_Price')
##df_average_price[['City of Edinburgh','City of Glasgow', 'Stirling','Orkney Islands']].plot()
df_average_price.plot()
# adding title to the plot
plt.title('Average price')
  
# adding labels to the axes, add a legend
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.legend()
```

```python
plt.figure(figsize=(16, 8), dpi=300)
df_annual_change=df_scotland.pivot(index='Date',columns='Region_Name',values='Annual_Change')
##df_annual_change[['City of Edinburgh','City of Glasgow', 'Stirling','Orkney Islands']].plot()
df_annual_change.plot()
# adding title to the plot
plt.title('')
  
# adding labels to the axes, add a legend
plt.xlabel('Date')
plt.ylabel('Annual Change')
plt.legend()
```

```python
plt.figure(figsize=(16, 8), dpi=300)
df_monthly_change=df_scotland.pivot(index='Date',columns='Region_Name',values='Monthly_Change')
#df_monthly_change[['Highland', 'Stirling', 'Orkney Islands' , 'City of Edinburgh']].plot()
df_monthly_change.plot()
plt.title('')
  
# adding labels to the axes, add a legend
plt.xlabel('Date')
plt.ylabel('Monthly Change')
plt.legend()
```
