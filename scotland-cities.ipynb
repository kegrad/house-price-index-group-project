```python
scotland20040101 = df[df['Area_Code'].str.match('S1') & df['Date'].str.match('2004-01-01')]
scotland20040101
```

```python
# based on https://www.python-graph-gallery.com/choropleth-map-geopandas-python
import geopandas as gpd
import geoplot as gplt

#geoData = gpd.read_file("https://github.com/martinjc/UK-GeoJSON/blob/master/json/administrative/sco/lad.json?raw=true")
geoData = gpd.read_file("lad.json")
```

```python
geoData
```

```python
fullData = geoData.merge(scotland20040101, left_on=['LAD13CD'], right_on=['Area_Code'])
fullData
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, 1, figsize=(16, 12))

# Set up the color sheme:
import mapclassify as mc
scheme = mc.Quantiles(fullData['Average_Price'], k=10)

# Map
gplt.choropleth(fullData, 
    hue="rate", 
    linewidth=.1,
    scheme=scheme, cmap='inferno_r',
    legend=True,
    edgecolor='black',
    ax=ax
);

ax.set_title('Average house prices in Scottish local authorities, January 2004', fontsize=13);
```
